## **Statistics**

Statistics is a fundamental pillar for extracting meaningful insights from complex datasets.It is used for :--

 1.Data analysis: Data scientists use statistics to analyze raw data, build models, and infer results.

 2.Pattern recognition: Data scientists use statistics to identify patterns and trends in data.

 3.Communication: Data scientists use statistics to communicate their findings to others.

 4.Decision making: Businesses and organizations use statistical data science to make decisions and evaluations.

```python
#Population
```

**Population**

Definition: The entire group of individuals or observations that you are interested in studying.

Characteristics:

Can be finite or infinite.

Includes every member that fits a set of specifications.

Examples:

All the residents of a country.

Every transaction made by a company in a year.

The complete set of data points generated by a sensor over its lifetime.

```python
#Sample
```

**Sample**

Definition: A subset of the population selected for analysis.

Characteristics:

Should represent the population accurately.

Easier and more cost-effective to collect than data from the entire population.

Examples:

1,000 residents randomly selected from a country's population survey.


A month's worth of transactions out of a company's yearly data.

Data collected from sensors during a specific period.


```python
# Descriptive
```

**Descriptive Statistics**

Definition: Descriptive statistics summarize and describe the features of a dataset.

Purpose: To provide a clear, concise overview of the data without drawing conclusions beyond the data itself.

Types:

 1. Measures of Central Tendency:
Mean, Median, Mode (summarize the center of the data).
 2. Measures of Spread:
Range, Variance, Standard Deviation, Interquartile Range (describe data variability).
 3. Other Summaries:
Frequency distributions, Percentiles, and Graphical representations like histograms and box plots.

Example: The average score of students in a class is 75, with a standard deviation of 5.

```python
# Inferential
```

**Inferential Statistics**

Definition: Inferential statistics make predictions or inferences about a population based on a sample of data.

Purpose: To generalize findings from a sample to the entire population.

Types:
 1. Estimation:
  Point Estimate (e.g., sample mean as an estimate of population mean) and
  Confidence Intervals.
 2. Hypothesis Testing:
 Used to test claims about a population (e.g., t-test, z-test, chi-square test).
 3. Regression Analysis:
 Predicting values of one variable based on another (e.g., linear regression).
 4. ANOVA:
 Comparing means across multiple groups.

Example: Based on a sample of 100 students, we estimate that the average score of all students in a school is between 70 and 80 with 95% confidence.

**Covariance and Correlations**

Covariance and correlation are both measures of the relationship between two variables:


1. Covariance indicates the direction of the linear relationship between two variables. If both variables tend to increase or decrease together, the covariance is positive. If one increases while the other decreases, it's negative. However, it doesn't indicate the strength of the relationship and is sensitive to the scale of the data.


2. Correlation standardizes covariance to measure both the strength and direction of the relationship on a fixed scale. It ranges from -1 to 1:

 +1 indicates a perfect positive linear relationship.

 -1 indicates a perfect negative linear relationship.

 0 means no linear relationship.

Correlation is more interpretable since it’s not influenced by the scale of the data.

**Central Limit Theorem**

The Central Limit Theorem (CLT) states that the distribution of the sample mean of a large enough sample size will approximate a normal distribution, regardless of the shape of the population distribution, provided the samples are independent and identically distributed.


Key points:

 1. As sample size increases, the sample mean becomes normally distributed.
 2. This applies even if the original data distribution is not normal.
 3. CLT is crucial for making inferences about population parameters from sample data.
