# **Introduction to Matplotlib**

Matplotlib is a powerful plotting library in Python used for creating static, animated, and interactive visualizations. Matplotlibâ€™s primary purpose is to provide users with the tools and functionality to represent data graphically, making it easier to analyze and understand. It was originally developed by John D. Hunter in 2003 and is now maintained by a large community of developers.

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

**Line chart**

Use for: Showing trends over time or continuous data.

Example: Stock prices over time.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [150, 200, 250, 300, 350]

# Plot
plt.plot(months, sales, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.grid(True)
plt.show()
```

Matplotlib Bar Plot

Use for: Comparing quantities across categories.

Example: Sales by product category.

```python
# Bar plot
x=["python","Java","c","c++"]
y=[90,80,50,75]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title("L vs P",fontsize=18)
c=["y","r","b","g"]
plt.bar(x,y,width=0.5,color=c,align="center",edgecolor="black",linewidth=2,alpha=0.8,label="bar plot")
plt.legend()
plt.show()
```

Matplotlib Scattor Plot

Use for: Exploring relationships between two numerical variables.

Example: Height vs. weight correlation.


```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
size=[500,200,234,233,600,300,290]
col=["g","b","y","r","g","y","b"]
plt.scatter(day,no,c=col,s=size,marker="*",edgecolor="black")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

**Matplotlib Pie Plot**

Use for: Showing part-to-whole relationships (few categories).

Example: Market share distribution.

```python
x=[10,20,30,40]
y=["C","C++","Java","Python"]
exclu=[0.2,0.01,0.02,0.1]
col=["g","b","y","m"]

plt.title("Pie Plot")
plt.pie(x,labels=y,explode=exclu,colors=col,autopct="%0.2f%%",shadow=True)
plt.legend(loc=2)
plt.show()
```

**Matplotlib Stem Plot**

Use for: Representing discrete changes or cumulative data.

Example: Cumulative sales over time.

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",bottom=3.5,basefmt="g",label="python")
plt.legend()
plt.show()
```

**Matplotlib Box Plot**

Use for: Summarizing distributions and detecting outliers.

Example: Exam scores across different classes.

```python
x=[10,20,30,40,50,60,70,130]

plt.boxplot(x,widths=0.2,labels=["python"],patch_artist=True)
plt.show()
```

**Matplotlib Stack Plot**

use:Part-to-whole relationships over time.

Cumulative data visualization.

Example:
Showing sales from multiple products over months.

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"])
plt.legend(loc=2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Python")
plt.show()
```

**Matplotlib Step Plot**

Use for: Representing discrete changes or cumulative data.

Example: Cumulative sales over time.

```python
x=[1,2,3,4,5]
y=[2,3,4,3,4]

plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="python")

plt.title("python")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc=2)
plt.grid()
plt.show()
```

**Fill_Between**

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4))

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.savefig("fill.jpg",dpi=2000)

plt.show()
```

**Subplot**

Use for: Displaying multiple plots in one figure for comparison.

Example: Combining line and scatter plots in one view.

```python
# Create the figure
plt.figure(figsize=(10, 8))

# First subplot (2 rows, 2 columns, 1st subplot)
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Plot 1')

# Second subplot (2 rows, 2 columns, 2nd subplot)
plt.subplot(2, 2, 2)
plt.bar(['A', 'B', 'C', 'D'], [3, 7, 8, 5])
plt.title('Plot 2')

# Third subplot (2 rows, 2 columns, 3rd subplot)
plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30], color='red')
plt.title('Plot 3')

# Fourth subplot (2 rows, 2 columns, 4th subplot)
plt.subplot(2, 2, 4)
plt.pie([30, 40, 20, 10], labels=['A', 'B', 'C', 'D'])
plt.title('Plot 4')

# Show the plot
plt.tight_layout()  # Automatically adjust spacing
plt.show()
```
