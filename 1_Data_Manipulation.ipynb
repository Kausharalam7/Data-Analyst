**Introduction to NumPy**

NumPy stands for Numerical Python, is an open-source library in Python used for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

```python
# installation of numpy
!pip install numpy
```

```python
# importing numpy
import numpy as np

# Creating a 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
# Creating a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
```

```python
# array attributes
print(arr2.ndim)    # Number of dimensions
print(arr2.shape)   # Shape of the array
print(arr2.size)    # Total number of elements
print(arr2.dtype)   # Data type of elements
print(arr2.itemsize) # Size of each element in bytes
print(arr2.nbytes)   # Total bytes consumed by the array
```

```python
# We can create NumPy arrays from Python lists or tuples using the array() function.

# from a list
list1=[1,2,3,4,5]
arr1=np.array(list1)
print(arr1)

# from a tuple
tuple1=(3,4,5,67,8)
arr2=np.array(tuple1)
print(arr2)
print(arr2.shape)
print(arr2.ndim)
print(arr2.size)
```

Using Built-in Functions

NumPy provides several functions to create arrays with specific patterns or values.


1. zeros(): Create an array filled with zeros.

2. ones(): Create an array filled with ones.

3. arange(): Create an array with a range of values.

4. linspace(): Create an array with a specified number of evenly spaced values.

5. eye(): Create an identity matrix.

```python
zeros_arr=np.zeros((2,3))
print(zeros_arr)

ones_arr=np.ones((2,4))
print(ones_arr)

range_arr=np.arange(1,10,2)
print(range_arr)

linspace_arr=np.linspace(0,10,5)  # Array with evenly spaced values
print(linspace_arr)

eye_arr=np.eye(3) # identity matrix
print(eye_arr)
```

```python
## Using random numbers
# random integers
rand_int=np.random.randint(0,10,size=(3,3))
print(rand_int)

## random floats
rand_float=np.random.random((2,3))
print(rand_float)
```

```python
# Array indexing and slicing
arr=np.array([11,22,33,44,55,66,77,88,99])
print(arr[0]) # first element
print(arr[-1]) # last element

print(arr[2:7])
print(arr[:5])
print(arr[5:])

# for multidimensional
arr3=np.array([[1,2,3],[4,5,6]])
print(arr3[0,2])
print(arr3[0,:])
print(arr3[:,2])
```

```python
## Boolean Indexing
arr5=np.array([10,20,30,40,50])
print(arr5[arr5>30])
print(arr5[(arr5>=20) & ( arr5<40)])

# for multidimensional arrays
arr6=np.array([[1,5,3,],[2,3,4],[3,5,7]])
print(arr6[arr6>2])
arr6[arr6 > 4] = 0
print(arr6)
```

```python
# array operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr2 / arr1)
```

```python
# Allows arithmetic operations between arrays of different shapes
arr = np.array([1, 2, 3])
scalar = 2
print(arr + scalar)
print(arr*scalar)

arr2 = np.array([[1, 2, 3], [4, 5, 6]])
scalar = 10
print(arr2 + scalar)
print(arr2*scalar)
```

```python
# Universal Function
arr=np.array([1,4,9,16,25,49,81,36,225])
print(np.sqrt(arr))
print(np.log(arr))
print(np.exp(arr))
print(np.tan(arr))
print(np.sin(arr))
print(np.cos(arr))
```

```python
# reshapings arrays
arr=np.arange(1,10)
reshaped=arr.reshape(3,3)
print(reshaped)
```

```python
# Flatten -- > Convert a multi-dimensional array into a 1D array.
arr=np.array([[2,3,4,5],[2,3,4,5]])
print(arr.flatten())
```

```python
# Transpose
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
transposed = arr2.T
print(transposed)
```

Stacking and Splitting Arrays

Stacking

Combine multiple arrays into one.


1. np.vstack(): Stack arrays vertically (row-wise).

2. np.hstack(): Stack arrays horizontally (column-wise).

3. np.concatenate(): General-purpose concatenation along an existing axis.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
v_stack = np.vstack((arr1, arr2))
print("Vertical Stack:\n", v_stack)
h_stack = np.hstack((arr1, arr2))
print("Horizontal Stack:", h_stack)
# Concatenate along axis 0
concat_axis0 = np.concatenate((arr1, arr2), axis=0)
print("Concatenate Axis 0:", concat_axis0)
# Concatenate along axis 1
arr3 = np.array([[7, 8, 9], [10, 11, 12]])
concat_axis1 = np.concatenate((arr1.reshape(1, -1), arr3), axis=0)
print("Concatenate Axis 0 with 2D Array:\n", concat_axis1)
```

Splitting

Divide an array into multiple sub-arrays.


1. np.split(): Split array into equal parts.

2. np.hsplit(): Split array horizontally.

3. np.vsplit(): Split array vertically.

```python
arr=np.arange(10,42)
#split into 3 equal parts
splits=np.split(arr,4)
for i in splits:
  print(i)

# Split a 2D array horizontally
arr2d = np.array([[1, 2, 3, 4,3,5], [5, 6, 7,4,5, 8]])
hsplits = np.hsplit(arr2d, 3)
for j in hsplits:
    print("Horizontal Split:\n", j)

# Split a 2D array vertically
vsplits = np.vsplit(arr2d, 2)
for k in vsplits:
    print("Vertical Split:\n", k)
```

Linear Alegebra with NumPy

```python
# Dot product
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
dot = np.dot(arr1, arr2)
print("Dot Product:", dot)

# For 2D arrays
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
dot_matrix = np.dot(matrix1, matrix2)
print("Dot Product of Matrices:\n", dot_matrix)
```

```python
# Matrix Operations
result=matrix1 @ matrix2
print(result)

transpose=matrix1.T
print(transpose)

# inverse of a matrix
inverse=np.linalg.inv(matrix1)
print(inverse)
```

```python
# Eigenvalues and Eigenvectors
matrix = np.array([[2, 0], [0, 3]])
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Setting a seed for reproducibility
np.random.seed(142)
print("Random Integers with Seed:\n", np.random.randint(0, 10, size=5))
```

```python
# Random Distributions

# Normal Disrtibution
normal=np.random.normal(loc=0,scale=1, size=5)
print(normal)

#Binomial Distribution
binomial=np.random.binomial(n=10,p=0.5,size=5)
print(binomial)

# Poisson distribution
poisson=np.random.poisson(lam=3,size=5)
print(poisson)
```

```python
# creating a Dataframe from  a dictionary
data={
    'Name':['Alice','Bob','Charlie','David'],
    'Age':[12,2,32,35],
    'City':['New York','Los Angeles','Chicago',"Patna"]
}
df=pd.DataFrame(data)
print(df)

# creating a dataframe with custom index
df_custom_index=pd.DataFrame(data,index=['a','b','d','f'])
print(df_custom_index)
```

Key Differences Between Series and DataFrame

Dimension:

1. Series: One-dimensional (like a single column).

2. DataFrame: Two-dimensional (like a table with rows and columns).

Structure:

1. Series: Can hold a single data type.
2. DataFrame: Can hold multiple data types (different types in different columns).

Usage:

1. Series: Useful for storing and manipulating single columns of data.
2. DataFrame: Useful for handling datasets with multiple attributes.
