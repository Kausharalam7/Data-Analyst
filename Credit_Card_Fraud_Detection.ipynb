Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Preprocessing

```python
card_dataset=pd.read_csv('/content/creditcard.csv')
```

```python
print(card_dataset)
```

Filling the missing value with the mean of given dataset

```python
card_dataset.fillna(card_dataset.mean(),inplace=True)
```

```python
card_dataset = card_dataset.astype(int)
```

```python
card_dataset.isnull().sum()
```

```python
# statistical view
card_dataset.describe()
```

```python
# cheking the distribution of label(target) values
card_dataset['Class'].value_counts()
```

Splitting the Features and target

```python
X=card_dataset.drop(columns='Class',axis=1)
Y=card_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MOdel Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score


```python
# accuarcy on training data
X_train_prediction=model.predict(X_train)
training_data_prediction=accuracy_score(Y_train,X_train_prediction)
```

```python
print("Accuracy on training data : ",training_data_prediction)
```

```python
# accuarcy on training data
X_test_prediction=model.predict(X_test)
testing_data_prediction=accuracy_score(Y_test,X_test_prediction)
```

```python
print("Accuracy on testing data : ",testing_data_prediction)
```
